```python
# get the world country names to filter out 
country = pd.read_csv('country_list.csv')
```

```python
# preprocess the country_list file
country.columns = ['Country Name']
country = country[country['Country Name'].str.contains(r'[a-zA-z]')]
country['Country Name']  = country['Country Name'].str.split('|')
country['Country Name'] = country['Country Name'].str.get(1)
country = country['Country Name'].str.upper()
df['Country Name'] = df['Country Name'].str.upper()
new_df = pd.merge(df,country,on='Country Name')
new_df['Country Name'].unique()
```

```python
new_df.head(10)
```

```python
# change the column year to numeric data type
new_df['Year'] = pd.to_numeric(new_df['Year'])
# only select year from 1990 to 2019
new_df = new_df[new_df['Year'].between(1990,2019)]
```

```python
# fill the null with mean vaue
new_df = new_df.fillna(new_df.mean(axis=0))
```
